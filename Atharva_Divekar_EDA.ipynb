**RFM** - https://medium.com/@hhuseyincosgun/customer-segmentation-rfm-analysis-recency-frequency-monetary-5b29d5d45e35

```python
customer_data = merged_df.groupby('CustomerID').agg(Recency=('TransactionDate', lambda x: (merged_df['TransactionDate'].max() - x.max()).days),
                                                    Frequency=('TransactionID', 'count'),
                                                    Monetary=('TotalValue', 'sum'))

customer_data['CustomerSegment'] = pd.qcut(customer_data['Monetary'], 4, labels=['Low', 'Medium', 'High', 'VIP'])
customer_data
```

```python
customer_data['R_Score'] = customer_data['R_Score'].astype(int)
customer_data['F_Score'] = customer_data['F_Score'].astype(int)
customer_data['M_Score'] = customer_data['M_Score'].astype(int)

rfm_heatmap = customer_data.groupby(['R_Score', 'F_Score']).agg({'M_Score': 'mean'}).reset_index()
rfm_pivot = rfm_heatmap.pivot(index='R_Score', columns='F_Score', values='M_Score')

plt.figure(figsize=(8, 6))
sns.heatmap(rfm_pivot, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': 'Monetary Score'})
plt.title('RFM Grid Heatmap (Mean Monetary Scores)', fontsize=14)
plt.xlabel('Frequency Score (F)', fontsize=12)
plt.ylabel('Recency Score (R)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
freq_categories = customer_data.groupby('F_Score').size()

freq_categories.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Customer Segmentation by Frequency')
plt.xlabel('Frequency Score (1 = Low, 4 = High)')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

```python
customer_segments = customer_data['CustomerSegment'].value_counts()

# Plot pie chart
customer_segments.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#FF9999', '#FFCC99', '#99CCFF', '#66B2B2'])
plt.title('Customer Segmentation by Monetary Value')
plt.ylabel('')
plt.show()
```

```python
customer_data['Churned'] = customer_data['Recency'].apply(lambda x: 'Yes' if x > 90 else 'No')
churn_summary = customer_data['Churned'].value_counts()
print(churn_summary)
```

```python
churned_customers = customer_data[customer_data['Churned'] == 'Yes']
print(churned_customers)
```

```python
segment_analysis = customer_data.groupby('CustomerSegment').agg(
    Avg_Recency=('Recency', 'mean'),
    Avg_Frequency=('Frequency', 'mean'),
    Total_Monetary=('Monetary', 'sum'),
    Count=('CustomerSegment', 'size')
)
print(segment_analysis)
```

```python
merged_df['FirstPurchase'] = merged_df.groupby('CustomerID')['TransactionDate'].transform('min')
merged_df['TimeToPurchase'] = (merged_df['FirstPurchase'] - merged_df['SignupDate']).dt.days
avg_time_to_purchase = merged_df.groupby('Region')['TimeToPurchase'].mean()
avg_time_to_purchase
```

```python
avg_time_to_purchase.plot(kind='bar', color='blue')
plt.xlabel('Region')
plt.ylabel('Average Time to First Purchase (days)')
plt.title('Average Time to First Purchase by Region')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better visibility
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='TimeToPurchase', data=merged_df, palette='Set2')
plt.xlabel('Region')
plt.ylabel('Time to First Purchase (days)')
plt.title('Distribution of Time to First Purchase by Region')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Region', y='TimeToPurchase', data=merged_df, palette='muted')
plt.xlabel('Region')
plt.ylabel('Time to First Purchase (days)')
plt.title('Violin Plot: Time to First Purchase by Region')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for region in merged_df['Region'].unique():
    sns.histplot(merged_df[merged_df['Region'] == region]['TimeToPurchase'], kde=True, label=region, bins=20)

plt.xlabel('Time to First Purchase (days)')
plt.ylabel('Frequency')
plt.title('Distribution of Time to First Purchase by Region')
plt.legend(title='Region')
plt.show()
```

```python
merged_df['YearMonth'] = merged_df['TransactionDate'].dt.to_period('M')
time_purchase_by_region_month = merged_df.groupby(['YearMonth', 'Region'])['TimeToPurchase'].mean().unstack()

plt.figure(figsize=(12, 6))
sns.heatmap(time_purchase_by_region_month, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5)
plt.xlabel('Region')
plt.ylabel('Month')
plt.title('Heatmap: Average Time to First Purchase by Region and Month')
plt.show()
```
